```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
m = 100
X = 6 * np.random.rand(m, 1) - 3
y = 0.5 * X**2 + X + 2 + np.random.randn(m, 1)
```

```python
plt.scatter(X, y, color='blue')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Non linear data')
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_features = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly_features.fit_transform(X)
```

```python
print(X[0])
print(X_poly[0])
```

X_poly contains the original X, as well as the square of X. Then we can fit a linear regression model to the extended data:

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X_poly, y)
print(lin_reg.intercept_, lin_reg.coef_)
```

```python
theta0 = 2.01275717
theta1 = 0.97238811
theta2 = 0.54815506
x_curve = np.linspace(-3, 3, 100)
y_curve = theta0 + theta1 * x_curve + theta2 * x_curve**2
plt.scatter(X, y, color='blue', label='Generated Data')
plt.plot(x_curve, y_curve, color='red', label='Curve with Given Coefficients')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Data and Curve')
plt.legend()
plt.show()
```
